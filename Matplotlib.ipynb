```python
#plotting using matplotlib
#plt scatter
plt.scatter(x,y,c='b')
plt.xlabel('X axis')    
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
plt.plot(x,y,'r*-')
```

```python
y=x*x
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'o',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.plot("Y axis")
plt.title('2D diagram')
```

```python
plt.subplot(2,2,1)
plt.plot(a,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
```

```python
x=np.arange(1,11)
y=7*x+5
plt.title('Matplotlib demo')
plt.xlabel('X axis caption')
plt.ylabel('Y axis caption')
plt.plot(x,y,'r--')
```

```python
x=np.arange(0,4*np.pi,0.1)
print(x)
y=np.sin(x)
print(y)
plt.title('sine wave form')
plt.plot(x,y)
```

```python
#computing sin and cos
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin,'r--')
plt.title('sine wave form')
plt.subplot(2,2,4)
plt.plot(x,y_cos,'g--')
plt.title('cosine wave form')
```

```python
#bar plot
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')  
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')      
```

```python
#histograms
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('histogram')
#lets us know the frequency of the data

```
